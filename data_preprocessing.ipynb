```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.under_sampling import RandomUnderSampler
import joblib
```

```python
data = pd.read_csv(r'E:\CoffeeCoders-EasyFundBuddy\dataset\new_loan.csv')
data.head(n=10)
```

# PreProcessing Starts

```python
data = data.drop(['nmi_emi_Ratio'], axis=1)

# Handle Dataset Imbalance

# Separate features and target
X = data.drop('LoanApproved', axis=1)  # replace 'target_column' with the actual target column name
y = data['LoanApproved']


# Apply RandomUnderSampler
rus = RandomUnderSampler(random_state=4)
X, y = rus.fit_resample(X, y)

# Add the target column back to the DataFrame
X['LoanApproved'] = y

X.to_csv(r'dataset/preprocessed_loan.csv', index=False)
```

```python
df = pd.read_csv(r'dataset/preprocessed_loan.csv')

# Encoding categorical data
label_encoders = {}
for col in data.select_dtypes(include=['object']).columns:
    label_encoders[col] = LabelEncoder()
    data[col] = label_encoders[col].fit_transform(data[col])
```
